This fascinating journey into derivatives is begun with the sleek and simple **circle**.
<br><br><br>
The equation for the *unit circle* is defined as:
$$
  x^2 + y^2 =1.
$$
Through algebraic manipulation, y is equated to:
$$
  y = \sqrt{1-x^2}
  \mbox{ and } y = - \sqrt{1-x^2}
$$


Using this representation of y, a circle can be plotted using the *numpy* and *matplotlib* Python libraries, as seen below.

````python
def f(x):
  return np.sqrt(1-x**2)

xs = np.linspace(-1, 1, 100)
ys = f(xs)

plt.figure(figsize=(10, 10), dpi=80)
plt.plot(xs, ys, color="black")
plt.plot(xs, -ys, color="black")
````

A derivative represents the rate of change of a function.<br><br>

Any function's derivative can be approximated using the following equations (any one could work; they are all simply different ways to express the same equation):

$$ f'(x) ≈ \frac{rate\, of\, change\, of\, f}{rate\, of\, change\, of\, x}  = \frac{f(x)- f(x - \delta)}{\delta} = \frac{f(x +\delta)- f(x - \delta)}{2\delta}= \frac{f(x +\delta)- f(x)}{\delta}  = \frac{f(x +\delta_1)- f(x-\delta_2)}{\delta_1 + \delta_2},$$

where δ is an extremely small decimal number, such as 0.00000001.<br><br>
This delta approximation of a function's derivative is also seen in the below function, called *df*.



````python
def df(f, x0):
  delta = 0.00000000001
  return  (f(x0+delta) - f(x0-delta))/(2*delta)
````

To find a precise *analytic* derivative, a set of rules, called derivative rules, are used. Choosing a particular one depends on the mathematical operations used in the function one is trying to differentiate.

When the derivative of a function is found, one can find the value of the tangent at any point *x*. Graphically, the tangent is represented as a vector, with the following equation:

$$ y = ax + b, $$

where x is a given point, a is the slope of the tangent vector (indicates the rate at which y changes with one-unit increase in x), and b is the y-intercept of the vector (the value of y when x = 0).

The value of a can be found through:

$$ a = \frac{dy}{dx}, $$

where, in this case, y is the function for which one is attempting to determine the tangent vector at a specific point *x*.

The value of b can be found through:

$$ b = f(x) - ax, $$

where a is the value found before and x is the same specified point as above.

By drawing the derivative at enough points of a function's graph, an outline of the function's form is seen:

````python
nsamples = 100
for _ in range(nsamples):
  x1 = np.random.random()
  y1 = f(x1)

  a = df(f, x1)
  b =  y1-a*x1

  epsilon = 0.1
  x = np.linspace(x1-epsilon, x1 +epsilon , 100)
  tangent = a*x + b
  plt.plot(x1, y1, 'o', color="magenta")
  plt.plot(x, tangent, color="pink")
````

Note that above, rather than having a particular value of *x*, we use a linspace, which is a lineraly spaced array of, in this case, 100 values, that for each the tangent line is found. This process of drawing the tangent vectors for 100 values is repeated through the use of a for loop, causing a cohesive depiction of the outline of a function's graph.

The range of the values in the linspace are based on a randomly picked value between 0.0 (inclusive) and 1.0 (exclusive). When the value is chosen, 100 values between the chosen value - epsilon (defined as 0.1) and the chosen value + epsilon are generated, creating the said linspace.

On a larger scale, one can also plot an entire circle using merely tangent vectors at tangent points on the graph of a circle:

````python
plt.figure(figsize=(10, 10), dpi=80)
# plt.plot(xs, ys, color="black")
# plt.plot(xs, -ys, color="black")

nsamples = 100
for _ in range(nsamples):
  x1 = np.random.random()
  y1 = f(x1)

  a = df(f, x1)
  b =  y1-a*x1

  epsilon = 0.1
  x = np.linspace(x1-epsilon, x1 +epsilon , 100)
  tangent = a*x + b
  plt.plot(x1, y1, 'o', color="magenta")
  plt.plot(x, tangent, color="pink")

for _ in range(nsamples):
  x1 = -np.random.random()
  y1 = f(x1)

  a = df(f, x1)
  b =  y1-a*x1

  x = np.linspace(x1-epsilon, x1 +epsilon , 100)
  tangent = a*x + b
  plt.plot(x1, y1, 'o', color="magenta")
  plt.plot(x, tangent, color="pink")

for _ in range(nsamples):
  x1 = -np.random.random()
  y1 = f(x1)

  a = df(f, x1)
  b =  y1-a*x1

  x = np.linspace(x1-epsilon, x1 +epsilon , 100)
  tangent = a*x + b
  plt.plot(x1, y1, 'o', color="magenta")
  plt.plot(x, tangent, color="pink")

for _ in range(nsamples):
  x1 = -np.random.random()
  y1 = -f(x1)

  a = df(f, x1)
  b =  y1-a*x1

  x = np.linspace(x1-epsilon, x1 +epsilon , 100)
  tangent = -(a*x + b)
  plt.plot(x1, y1, 'o', color="magenta")
  plt.plot(x, tangent, color="pink")

for _ in range(nsamples):
  x1 = np.random.random()
  y1 = -f(x1)

  a = df(f, x1)
  b =  y1-a*x1

  x = np.linspace(x1-epsilon, x1 +epsilon , 100)
  tangent = -(a*x + b)
  plt.plot(x1, y1, 'o', color="magenta")
  plt.plot(x, tangent, color="pink")
````

Collectively,  all tangent vectors give this so-called *tangent envelope*.  Thus, by using 1-dimensional data, which are tangent vectors, one can recover the *unseen* orignal object, which is, in this case, the 2-dimensional circle. One can imagine doing the same for 4-dimensional objects, which are objects in time and space. In such an instance, one would use the notion of the tangent envelope (which uses data that is one dimension less than the original object, which is 3-dimensional data in this case) to capture the original 4-dimensional object, which is otherwise impossible to observe directly with the naked human eye.
